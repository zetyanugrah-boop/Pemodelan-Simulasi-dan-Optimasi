#1. Gunakan distribusi Poisson untuk mensimulasikan jumlah pelanggan dalam satu jam (60 menit)

```python
import numpy as np

# rata-rata pelanggan per jam
lam = 30

# simulasi 1 jam
np.random.seed(42)   # agar hasil tetap sama
pelanggan_1_jam = np.random.poisson(lam=lam)

print("Jumlah pelanggan dalam 1 jam:", pelanggan_1_jam)
```

#Interpretasi
Karena rata-rata pelanggan per jam adalah 30, maka simulasi jumlah pelanggan dalam 1 jam dapat dilakukan dengan:

                                 X∼Poisson(30)
lam = 30 karena rata-rata 30 pelanggan/jam.

##### np.random.poisson(lam=30) menghasilkan 1 angka acak jumlah pelanggan dalam 1 jam.
##### Berdasarkan soal, rata-rata kedatangan pelanggan adalah 5 orang setiap 10 menit. Karena dalam 1 jam terdapat 6 interval 10 menit, maka rata-rata jumlah pelanggan yang datang dalam 1 jam adalah 30 orang. Oleh karena itu, jumlah pelanggan per jam dapat dimodelkan menggunakan distribusi Poisson dengan parameter 𝜆=30. Distribusi ini sesuai karena digunakan untuk memodelkan banyaknya kejadian (kedatangan pelanggan) dalam suatu interval waktu tertentu.


#2. Jalankan simulasi untuk 100 hari dan visualisasikan hasilnya dalam histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# parameter Poisson
lam = 30

# agar hasil konsisten
np.random.seed(42)

# simulasi 100 hari
simulasi_100_hari = np.random.poisson(lam=lam, size=100)

# tampilkan hasil simulasi
print("Hasil simulasi 100 hari:")
print(simulasi_100_hari)

# histogram
plt.hist(simulasi_100_hari, bins=10, edgecolor='black')
plt.title("Histogram Jumlah Pelanggan per Jam (100 Hari)")
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Frekuensi")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

#Interpretasi
Hasil simulasi selama 100 hari menunjukkan bahwa jumlah pelanggan per jam berfluktuasi di sekitar nilai rata-ratanya. Dari simulasi tersebut, diperoleh rata-rata sekitar 30 pelanggan per jam dan variansi juga mendekati 30. Hal ini menunjukkan bahwa hasil simulasi sudah sesuai dengan sifat distribusi Poisson, di mana nilai rata-rata dan variansi secara teori adalah sama dengan parameter
𝜆
λ. Dengan demikian, simulasi yang dilakukan dapat dikatakan mewakili kondisi nyata secara cukup baik.

#3. Hitung rata-rata dan variansi jumlah pelanggan per jam berdasarkan hasil simulasi

```python
# hitung rata-rata
rata_rata = np.mean(simulasi_100_hari)

# hitung variansi sampel
variansi = np.var(simulasi_100_hari, ddof=1)

print("Rata-rata hasil simulasi :", rata_rata)
print("Variansi hasil simulasi  :", variansi)
```

#Interpretasi
Dari simulasi 100 hari (dengan seed = 42), diperoleh contoh hasil:

Rata-rata simulasi ≈ 30.05

Variansi simulasi ≈ 29.24

Rata-rata hasil simulasi mendekati 30, sesuai dengan nilai teoritis distribusi Poisson.
Variansi juga mendekati 30, karena pada distribusi Poisson berlaku:

𝐸(𝑋)=𝜆

Var(X)=λ

Jadi secara teori:

𝐸(𝑋)=30,  𝑉𝑎𝑟(𝑋)=30

Berdasarkan histogram, sebagian besar jumlah pelanggan berada di sekitar 25 sampai 35 pelanggan per jam, dengan puncak frekuensi berada di sekitar nilai rata-rata. Pola ini menunjukkan bahwa kedatangan pelanggan cenderung terkonsentrasi di sekitar nilai tengah, sementara nilai yang terlalu kecil atau terlalu besar terjadi lebih jarang. Bentuk histogram yang menyerupai distribusi lonceng namun tetap diskrit juga mendukung bahwa data hasil simulasi mengikuti karakteristik distribusi Poisson.


#4. Jika restoran memiliki kapasitas maksimal 40 pelanggan per jam, berapa probabilitas terjadinya kelebihan kapasitas?

```python
from scipy.stats import poisson

# probabilitas kelebihan kapasitas
prob_kelebihan = 1 - poisson.cdf(40, mu=30)

print("Probabilitas kelebihan kapasitas:", prob_kelebihan)
```

#Interpretasi
Kelebihan kapasitas berarti:

 X > 40

Dengan:

X∼Poisson(30)

Maka yang dicari:

P(X>40)=1−P(X≤40)

####Hasil Perhitungan

Probabilitasnya adalah:

P(X>40)≈0.0323

atau sekitar:              3.23%

Artinya, peluang jumlah pelanggan melebihi 40 orang dalam 1 jam adalah sekitar:         3.23%
	​

Jadi kemungkinan restoran kelebihan kapasitas relatif kecil, tetapi tetap ada sekitar 3 kali dalam 100 jam pengamatan.

Untuk kapasitas maksimal restoran sebesar 40 pelanggan per jam, diperoleh probabilitas kelebihan kapasitas sebesar sekitar 3,23%. Artinya, kemungkinan jumlah pelanggan yang datang melebihi 40 orang dalam satu jam tergolong kecil, yaitu hanya sekitar 3 dari 100 jam pengamatan. Meskipun peluangnya tidak besar, restoran tetap perlu mempertimbangkan langkah antisipasi pada jam-jam sibuk agar pelayanan tetap optimal ketika terjadi lonjakan pelanggan.
